### <font color = green|black> **Dados Históricos**</font>

Para extrairmos os dados financeiros, integrando a biblioteca pandas_datareader com o Alpaha Vantage, precisamos apenas utilizar a seguinte linha de código.
**web.DataReader("nome do ativo", "função", start = "início do período", end="fim do período")**

### <font color = green|black> Algumas observações: </font>

**Nome do ativo**: Quando for extrair dados de ações americanas, basta colocar como argumento o próprio ticker da ação.
Entretanto, ao procurar ações brasileiras, a nomenclatura especifica é o ticker da ação mais **".SAO"**.
Então, por exemplo, no caso da ação do Itaú, que possui o ticker ITUB4 e ITUB3, no código teria que escrever **ITUB4.SAO e ITUB3.SAO**.
O mesmo vale para o próprio índice Ibovespa, sua nomenclatura é "**BOVA11.SAO**".

**Start e End**: Deve-se colocar o início e o final do período, respectivamente. O formato a ser adotado pela medida temporal é datetime(ano, mês, dia). Se optar por não colocar nenhuma das informações referentes ao período analisado, o output será as informações de toda cronologia disponível.

**Tipo de output**: DataFrame.

### <font color = green|black> **Como interpretar os seguintes dados:** </font>
- **Open**: valor na abertura do mercado

- **High**: valor máximo atingido

- **Low**: valor minimo atingido

- **Close**: valor de fechamento

- **Adjusted close**: valor de fechamento ajustado

- **Volume**: quantidade transacionada

- **Dividend amount**: quantidade de dividendo

- **Split coefficient**: coeficiente de divisão

**Observação**:


**1°)** O **coeficiente de divisão** é obtido de acordo com a proporção adotada pela empresa quando é realizado o desdobramento de suas ações.
Lembrando que o desdobramento é a ação que as empresas da bolsa tomam para dividir suas ações em um número maior de ações. Desta forma, ações que possuem cotações elevadas passam a se tornar mais acessíveis ao público, sem que haja mudança no capital social. Por exemplo, uma ação, no valor de 20, desdobrada numa proporção de 1 para 2 terá sua cotação atualizada para 10 e, consequentemente, um coeficiente no valor de 2. Além disso, o coeficiente é responsável por atualizar o valor ajustado.

**2°)** A diferença entre o **fechamento** e o **fechamento ajustado** é que o segundo é muito mais útil para examinar os resultados históricos de uma empresa, pois os seus valores são ajustados toda vez que ocorrem divisões de ações (desdobramento) e dividendos. 

### <font color  = green|black> Vou listar a seguir as funções que podem ser utilizadas: </font>
- **av-intraday** - Retorna valores de ordens de Daytrade com open, high, low, close, volume 

- **av-daily** - Retorna valores diários de open, high, low, close, volume

- **av-daily-adjusted** - Retorna valores diários de open, high, low, close, volume, adjusted close, volume, dividend amount, split coefficient

- **av-weekly** - Retorna valores semanais open, high, low, close, volume

- **av-weekly-adjusted** - Retorna valores semanais de open, high, low, close, volume, adjusted close, volume, dividend amount, split coefficient

- **av-monthly** - Retorna valores mensais de open, high, low, close, volum

- **av-monthly-adjusted** - Retorna valores mensais de open, high, low, close, volume, adjusted close, volume, dividend amount, split coefficient

- **av-forex-daily** - Retorna valores do Mercado Cambial
